```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
covid19 = pd.read_csv('covid19_Confirmed_dataset.csv')
covid19.head()
```

# Removing un-necessary columns

```python
covid19.drop(['Lat','Long'],axis = 1,inplace = True)
```

```python
covid19.head(15)
```

# Aggrigating the covid dataset and grouping it by 'Country/Region'

As we can see for Australia different data row are present for different Province/State hence we need to add all the cases for each Province/State for every country.

```python
corona = covid19.groupby('Country/Region').sum()
```

```python
corona.head(10)
```

# Now we need to add a new column max_infection_rate

we need to get max infection for each country for every 24 hours and then add this as a column in the corona dataframe

```python
countries = list(corona.index)
infection_rate = []
for country in countries:
    infection_rate.append(corona.loc[country].diff().max())
corona['max_infection_rate'] = infection_rate
```

```python
corona.head()
```

Now as we have the max_infection_rate for each countries, we do not require the rest positive cases on each dates. Hence we will drop all except the max_infection_rate

```python
covid = pd.DataFrame(corona['max_infection_rate'])
covid.head()
```

Now importing World happiness report dataset

```python
whr = pd.read_csv('worldwide_happiness_report.csv')
whr.head()
```

Now removing columns that are not required for this analysis

```python
whr.drop(['Overall rank','Score','Generosity','Perceptions of corruption'],axis=1,inplace = True)
whr.head()
```

Now we are required to join the two dataframes and to ease this step we set the 'Country or region' as index

```python
whr.set_index(['Country or region'],inplace = True)
whr.head()
```

now before joining the two dataframe it is agood practice to check the shape of both the dataframes

```python
covid.shape
```

covid - 187 countries and 1 max_infection_rate

```python
whr.shape
```

whr - 156 countries and 4 cols(GDP per capita, Social support, Healthy life expectancy, Freedom to make life choices)

```python
final_df = covid.join(whr,how = 'inner')
final_df.head(10)
```
